```python
import nltk
from nltk.corpus import stopwords
from copy import deepcopy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import speech_recognition as sr

from bokeh.plotting import figure, output_notebook, show


import pymorphy2


class Word2Vec:
    def __init__(self, data):
        import gensim

        self.model = gensim.models.Word2Vec(self.preprocess_data(data), 
                                            min_count=1, size=50, workers=3, 
                                            window=3, sg=1)
        self.data = data
        self.stopwords = stopwords.words('russian')
        self.morph = pymorphy2.MorphAnalyzer()
        
        self.threasholds = {
            'positive': 0.8,
            'negative': 0.8,
        }
        
        self.positive = self.data[self.data['pstv'] > self.threasholds['positive']]['term']
        self.negative = self.data[self.data['ngtv'] > self.threasholds['negative']]['term']
        
        self.positive = list(self.positive)
        self.negative = list(self.negative)
        
        print(len(self.positive))
        print(len(self.negative))
    
    def preprocess(self, words):
        return self.lemmatize(self.delete_stopwords(words))
    
    def delete_stopwords(self, words):
        words = list(filter(lambda x: x.lower() not in self.stopwords,
                           map(lambda x: x.lower(), words.split())))
        return text
    
    def lemmatize(self, words):
        result = []
        for word in words:
            p = morph.parse(word)[0]
            result.append(p.normal_form)
        return result
    
    def similarity(self, word, *words):
        average = 0
        for word_ in words:
            try:
                average += self.model.similarity(word, word_)
            except:
                pass
            
        return average / len(words)
    
    def estimate_word(self, word):
        positive = self.similarity(word, *self.positive)
        negative = self.similarity(word, *self.negative)
        return (positive, negative)
    
    def estimate_words(self, words):
        estimation = []
        #words = self.preprocess(words)
        for word in words.lower().split():
            estimation.append(self.estimate_word(word)[0])
        return sum(estimation)
    
    def preprocess_data(self, data):
        data = deepcopy(data.apply(lambda x: ','.join(x.astype(str)), axis=1))
        data = pd.DataFrame({'clean': data})
        data = [row.split(',') for row in data['clean']]
        return data
    

import time
from datetime import datetime as dt


output_notebook()
p = figure(tools="pan,box_zoom,reset,save",
           y_axis_type="log", y_range=[0.001, 10**11], title="log axis example",
           x_axis_label='sections', y_axis_label='particles'
          )

class SpeechRegocnizer:
    def __init__(self, data):
        self.language = 'ru'
        self.speech_recognizer = sr.Recognizer()
        self.microphone = sr.Microphone()
        self.microphones = self.microphone.list_microphone_names()
        
        self.text = []
        self.estimations = []
        self.timesteps = []
        
        self.words = {}
        self.frequency = {}
        self.vocabulary = set()
        self.num_words = 0
        
        self.fucking_words = {}
        self.fucking_frequency = {}
        self.fucking_num_words = 0
        self.fucking_list = []

        self.word2vec = Word2Vec(data)

    def recognize(self, timestep):
        audio = self.get_microphone_input(timestep)
        try:
            text = self.speech_recognizer.recognize_google(audio, language=self.language)
            return text
        except:
            pass
        return ''
        
    def get_microphone_input(self, timestep):
        with self.microphone as sourse:
            self.speech_recognizer.adjust_for_ambient_noise(sourse)
            audio_text = self.speech_recognizer.record(sourse, duration=timestep)
        return audio_text
    
    def activate(self, timestep=3):
        while True:
            words = self.recognize(timestep)
            
            self.num_words += len(words.split())
            self.fucking_num_words += len([word for word in words.split() if '*' in word])
            self.fucking_list.append(self.fucking_num_words)
            
            self.add_new_words(words)
            
            print((words, str(dt.now())))
            estimation = self.word2vec.estimate_words(words)
            self.timesteps.append(dt.now())
            self.estimations.append(estimation)
            
            plt.plot(self.estimations)
            plt.pause(0.1)
                        
            command = words.lower()
            if 'команда стоп' in command:
                break
            if 'команда построить график' in command:
                None
            if 'словарь' in command:
                print(self.vocabulary)
        
    def add_new_word(self, word):
        if '*' not in word:
            if word in self.words.keys():
                self.words[word] += 1
            else:
                self.words[word] = 1
            self.vocabulary.add(word)
            self.frequency[word] = self.words[word] / self.num_words
        else:
            if word in self.fucking_words.keys():
                self.fucking_words[word] += 1
            else:
                self.fucking_words[word] = 1
            self.fucking_frequency[word] = self.fucking_words[word] /\
                                           self.fucking_num_words
            
    def add_new_words(self, words):
        words_ = words.lower().split()
        for word_ in words_:
            self.add_new_word(word_)
    

sentiment_data = pd.read_csv('sentiment.csv', sep=';')
speech_recognizer = SpeechRegocnizer(sentiment_data)
speech_recognizer.activate()

```
